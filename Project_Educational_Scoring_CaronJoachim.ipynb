```python
# Function to reduce memory usage:
def reduce_memory_usage(df):
    # Print memory usage before changing datatypes
    start_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage before: {:.2f} MB'.format(start_mem))
    
    # Change datatypes
    for col in df.columns:
        col_type = df[col].dtype
        
        # Change datatypes only for integers or floats
        if col_type != 'object':
            c_min = df[col].min()
            c_max = df[col].max()
            
            # Change datatypes for integers
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
                    
            # Change datatypes for floats
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
    
    # Print memory usage after changing datatypes
    end_mem = df.memory_usage().sum() / 1024**2 
    print('Memory usage now : {:.2f} MB'.format(end_mem)) 
    print('Memory usage decreased by {:.1f}%'.format(100 * (start_mem - end_mem) / start_mem))
```

```python
# Define a function to create a dataframe for each indicator
def create_indicator_dataframe(indicator_name, year, new_column):
    # keep only the rows concerning the mentioned indicator
    df = regions_indicators.loc[regions_indicators['Indicator Name'] == indicator_name]
    
    # Propagate the last valid observation to the last column in order to keep the most recent value
    df = df.fillna(method='ffill', axis=1)
    
    # Select the most recent value
    df = df[['Region', year]]
    
    # Rename the 'year' column with the indicator name
    df = df.rename(columns={year:new_column})
    
    display(df)
    return df
```

### SCORING for countries

```python
# SCORING
countries_sel_indicators['SCORE_GNI'] = (countries_sel_indicators['GNI'] 
                                         / countries_sel_indicators['GNI'].max())
countries_sel_indicators['SCORE_INTERNET'] = (countries_sel_indicators['Internet users (%)'] 
                                              / countries_sel_indicators['Internet users (%)'].max())
countries_sel_indicators['SCORE_ENROLMENT'] = (countries_sel_indicators['Enrolment upper secondary and tertiary'] 
                                               / countries_sel_indicators['Enrolment upper secondary and tertiary'].max())
countries_sel_indicators['SCORE_PRIVATE'] = (countries_sel_indicators['Private upper secondary and tertiary'] 
                                             / countries_sel_indicators['Private upper secondary and tertiary'].max())

# Synthesis Score
countries_sel_indicators['SCORE_SYNT'] = (countries_sel_indicators['SCORE_GNI']
                                          + countries_sel_indicators['SCORE_INTERNET']
                                          + countries_sel_indicators['SCORE_ENROLMENT']
                                          + countries_sel_indicators['SCORE_PRIVATE']) / 4
```
